```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt 
import seaborn as sns
%matplotlib inline
sns.set(color_codes=True)
```

```python
car_data=pd.read_csv('Automobile.csv')
```

```python
car_data.head()
```

```python
car_data.info()
```

# plotting univariate distributons

```python
sns.distplot(car_data['price'])
```

```python
sns.distplot(car_data['price'],kde=False)
```

```python
sns.distplot(car_data['price'],rug='True',color='g')
```

# bivariate distributons plotting 

```python
plt.scatter(car_data['price'],car_data['engine_size'])
```

```python
sns.jointplot(car_data['price'],car_data['engine_size'])
```

```python
sns.jointplot(car_data['engine_size'],car_data['horsepower'],color='r')
```

# hexagon bin plot

```python
sns.jointplot(car_data['engine_size'],car_data['horsepower'],kind='hex')
```

# kernal density estimator

```python
sns.jointplot(car_data['engine_size'],car_data['horsepower'],kind='kde')
```

# pairwise relationship in a data set

```python
sns.pairplot(car_data[['price','engine_size','horsepower']],diag_kind='kde')
```

```python
sns.pairplot(car_data[['price','engine_size','horsepower']])
```

# Plotting with categorical data

```python
sns.stripplot(car_data['engine_location'],car_data['price'])
```

```python
sns.stripplot(car_data['engine_location'],car_data['price'],jitter=False)
```

```python
sns.stripplot(car_data['fuel_type'], car_data['horsepower'])
```

```python
sns.swarmplot(car_data['fuel_type'],car_data['horsepower'])
```

```python
sns.swarmplot(car_data['fuel_type'], car_data['price'],hue=car_data['engine_location'])
```

# boxplot

```python
sns.boxplot(car_data['body_style'],car_data['price'],hue=car_data['fuel_type'])
```

```python
sns.boxplot(car_data['number_of_doors'],car_data['price'],hue=car_data['engine_location'])
```

# Bar plot

```python
sns.barplot(car_data['number_of_doors'],car_data['price'],hue=car_data['engine_location'])
```

```python
sns.countplot(x='number_of_doors',data=car_data)
```

# Drawing multi-panel categorical plot

```python
sns.factorplot('fuel_type','horsepower',hue='number_of_doors',col='engine_location',kind='swarm',data=car_data)
```

# visualize linear regression model

```python
sns.lmplot('horsepower','peak_rpm',data=car_data)
```

```python
sns.lmplot('horsepower','peak_rpm',data=car_data,hue='fuel_type')
```

```python
sns.lmplot('horsepower','peak_rpm',data=car_data,hue='engine_location')
```
